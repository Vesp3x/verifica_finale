```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("vehicles_dataset.csv")
```

```python
risultato = (
    df.groupby("manufacturer")
      .agg(
          modelli_unici=("model", "nunique"),
          count_modelli=("model", "count")
      )
      .reset_index()
)

print(risultato)
```

```python
x = np.arange(len(risultato["manufacturer"]))
width = 0.4

plt.figure(figsize=(12,6))

plt.bar(
    x - width/2,
    risultato["modelli_unici"],
    width=width,
    color="#1f77b4",
    label="Modelli unici"
)

plt.bar(
    x + width/2,
    risultato["count_modelli"],
    width=width,
    color="#ff7f0e",
    label="Totale modelli"
)

plt.xticks(x, risultato["manufacturer"], rotation=45, ha="right") 
plt.ylabel("Conteggio")
plt.yscale("log")
plt.title("Modelli per Manufacturer")

plt.legend()

plt.tight_layout()  
plt.show()
```
